```python
import csv

id_to_genesymbol = {}
with open ("fb_synonym_fb_2014_05.tsv") as f1:
    for line in f1:
        if line.startswith("FB"):
            line.rstrip()
            fb_id = line.split("\t")[0]
            gene_symb = line.split("\t")[1]
            id_to_genesymbol[fb_id] = gene_symb
            

outfile = open("FlyRNAi_data_baseline_vs_EGF_GO.txt","w") 
file_writer = csv.writer(outfile, delimiter=',')
with open ("FlyRNAi_data_baseline_vs_EGF.txt") as f2:
    next(f2)
    for line in f2:
        line.rstrip()
        fb_id1 = line.split()[0]
        baseline = line.split()[1]
        stimulus = line.split()[2]
        if fb_id1 in id_to_genesymbol:
            file_writer.writerow([id_to_genesymbol[fb_id1], baseline, stimulus])         
outfile.close()

gs_to_go = {}    
with open ("gene_association.goa_fly") as f3:
    for line in f3:
        if line.startswith("U"):
            line.rstrip()
            gs = line.split()[2]
            go = line.split()[3]
            gs_to_go[gs] = go
        
# for i in gs_to_go:
#     print (i, gs_to_go[i])
        
        
with open ("FlyRNAi_data_baseline_vs_EGF_GO.txt") as f4:
    for line in f4:
        line.strip()
        gene_s = line.strip(",")[0]
        baseline = line.split(",")[1]
        stimulus = line.split(",")[2]
        if gene_s in gs_to_go:
            print(gs_to_go[gs], baseline, stimulus)
        
```
